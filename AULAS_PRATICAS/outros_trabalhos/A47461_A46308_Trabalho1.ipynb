Gere o sinal $x(t)$ em Python considerando para o intervalo de tempo $0 < t < 1\,ms$, considerando que a frequência de amostragem é de 48KHz. Apresente numa figura o sinal no domínio do tempo e noutra figura o seu espetro.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Frequência
Fs = 48000
#Período
Ts = 1/Fs
# x entre 0 e 1
#começa no 0, acaba no 1, Ts = espaçamento entre valores
t = np.arange(0,0.001,Ts)
```

```python
#gráfico em cos
a = 20000 * np.cos(2*np.pi*5025*t)
#gráficos em sen
b = 10000 * np.sin(2*np.pi*2507*t)
#x1 é soma dos gráficos a e b
x1 = a + b
plt.plot(t,x1)
plt.title("a(t)")
plt.xlabel("t")
plt.ylabel("x")
plt.legend(["x1"])
plt.show()
```

```python
#np.fft.fft faz a transformada de fourier
X1 = np.fft.fft(x1)/len(x1) #coeficientes

#Retorne as frequências de amostra da Transformada Discreta de Fourier.
f = np.fft.fftfreq(len(x1), 1/Fs)

#Calcula o valor absoluto em termos de elementos.
amplitude = np.abs(X1)

#Retorne o ângulo do argumento complexo.
fase = np.angle(X1)

plt.stem(f, amplitude)
plt.title("a(t)")
plt.ylabel("Amplitude")
plt.xlabel("Frequência (Hz)")
plt.axis([-9500,9500,0,10000])
plt.show()

plt.stem(f, fase)
plt.title("a(t)")
plt.ylabel("Fase")
plt.xlabel("Frequência (Hz)")
plt.axis([-12000,12000,-3.5,3.5])
plt.show()
```
